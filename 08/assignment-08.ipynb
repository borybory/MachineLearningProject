##Computing Area

```
# Required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```
# Upload data
data = np.loadtxt('/content/drive/My Drive/Colab Notebooks/MachineLearningProject/08/data-pca.txt', delimiter=',')

x = data[:,0]
y = data[:,1]
```

```
def normalize_data(x, y):
    xn = (x-np.mean(x)) / np.std(x)
    yn = (y-np.mean(y)) / np.std(y)
    return xn, yn
```

```
def compute_covariance(x, y):
    x_tmp = x.reshape(1, len(x))
    y_tmp = y.reshape(1, len(y))
    z = np.concatenate([x_tmp,y_tmp])
    covar = np.dot(z, z.T) / x.shape[0]
    return covar
```

```
## Example results
plt.figure(figsize=(4,4))
plt.legend()
plt.title('original data points')
plt.scatter(x, y, marker='+', c='r')
```

```

xn, yn = normalize_data(x, y)
```

```
## Example results2

plt.figure(figsize=(5,5))
plt.legend()
plt.title('data nomalized by z-scoring')
plt.scatter(xn, yn, marker='+', c='r')
```

```
covar = compute_covariance(x, y)
print(covar)
```
